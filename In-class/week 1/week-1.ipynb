```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt

# Numpy
data = np.array([1, 2, 3, 4, 5], dtype=int)
print(data)
print(data[0])
print(data[0:2])

# Pandas
data = [
    {"Name": "Alice", "Age": 30, "City": "New York"},
    {"Name": "Bob", "Age": 25, "City": "Paris"},
    {"Name": "Charlie", "Age": 35, "City": "London"}
]
data = pd.DataFrame(data)
data["Age"] = data["Age"].astype(str)
print(data["Name"] + " is " + data["Age"] + " years old")

# Matplotlib
data_frame = pd.DataFrame({
    "Duration": [50, 40, 60, 45, 30],
    "Speed": [20, 25, 30, 35, 40]
})
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig, ax = plt.subplots(2, 2, figsize=(10, 8))


ax[0, 0].plot(x, y)
ax[0, 1].plot(data_frame["Duration"], data_frame["Speed"])
ax[1, 0].scatter(data_frame["Duration"], data_frame["Speed"])
ax[1, 1].bar(data_frame["Duration"], data_frame["Speed"])

ax[1, 1].pie(data_frame["Speed"], labels=data_frame["Duration"], autopct='%1.1f%%')

plt.show()




```
