```python
import numpy as np , pandas as pd , matplotlib.pyplot as plt
from sklearn. linear_model import LinearRegression
from sklearn import linear_model
 
# data = pd.read_csv("./Salary_Train_Data.csv")

# print(data.head())
# print(data.shape)

# data.plot(x='YearsExperience', y='Salary')
model = LinearRegression(fit_intercept=True)

test_data = pd.read_csv("./test_data.csv")
test_data.plot(x='Grade', y='CGPA')
 
X_train = test_data.drop('Grade', axis='columns')
Y_train = test_data.drop('CGPA' , axis='columns')
model.fit(X_train, Y_train)
print(model.intercept_) 
print(model.coef_)

Y_Predict = model.predict(X_train)

plt.scatter(X_train, Y_train, color='red')
plt.scatter(X_train, Y_Predict, color='blue')
```
