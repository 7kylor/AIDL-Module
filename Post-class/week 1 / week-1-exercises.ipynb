```python
# Week 1 In class slides exercises:

# 1. Create a data frame as follows
# age specialization cgpa
#  age cgpa  specialization

import pandas as pd, matplotlib.pyplot as plt ,seaborn as sns

data = {'age': [20, 21, 22, 23, 24, 19, 25, 19],
        'cgpa': [8.0, 7.5, 7.0, 7.5, 8.0, 7.5, 7.5, 8.0],
        'specialization': ['AI', 'ECE', 'Mech', 'AI', 'Logistics', 'IT', 'ECE', 'AI']}
df = pd.DataFrame(data)
print(df)

# 2. Plot a scatter plot with age on X - axis and gpa on Y-axis (use
# matplotlib and seaborn)

# using matplotlib
plt.scatter(df['age'], df['cgpa'])
plt.xlabel('Age')
plt.ylabel('CGPA')
plt.title('Age vs CGPA')
plt.show() 
 
# using seaborn
sns.set_theme(style="darkgrid")
sns.scatterplot(x='age', y='cgpa', data=df , legend='full')



# 3. Plot a scatter plot with age on X - axis and gpa on Y-axis (use
# seaborn)
sns.scatterplot(x='age', y='cgpa', data=df , legend='full' )
```
